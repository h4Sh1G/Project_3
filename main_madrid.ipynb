```python
#adding a column for the quarter
madrid_listing_1['quarter'] = 'Q1_2024'
madrid_listing_2['quarter'] = 'Q4_2023'
madrid_listing_3['quarter'] = 'Q3_2023'
madrid_listing_4['quarter'] = 'Q2_2023'
```

```python
#concatenating the data
madrid_df = pd.concat([madrid_listing_1, madrid_listing_2, madrid_listing_3, madrid_listing_4 ],ignore_index=True, axis=0)
```

```python
madrid_df.head(10)
```

```python
#removing unnecessary columns
columns_to_remove = ['last_review',
                   'number_of_reviews_ltm',
                   'license',
                   'neighbourhood_group']
                    #host name
madrid_df.drop(columns=columns_to_remove, inplace=True)
```

```python
#fill missing price values with mean
madrid_df['price'] = madrid_df['price'].fillna(madrid_df['price'].mean())
```
